```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from scipy.integrate import quad, dblquad
```

```python
import time
```

```python
def fs(x,v0,L):
    #return (v0/2.0)*(1 + np.cos((2*np.pi*x)/L))
    return v0*(1+np.sin((4*np.pi*x)/L))
```

```python
def fs_grad(x,v0,L):
    #return -(np.pi*v0/L)*np.sin((2*np.pi*x)/L)
    return ((4*v0*np.pi)/L)*np.cos((4*np.pi*x)/L)
```

```python

#tt = np.linspace(0,10000,100000)
v0 = 3.0
vw = 0.0
L = 20.0
xx = np.linspace(0.0,L,999)

plt.plot(xx,fs(xx,v0,L),label='fs')
plt.plot(xx,fs_grad(xx,v0,L),label='grad fs')
plt.legend()
```

```python
def chain(N):
    con_mat = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            if i==j:
                if i==0 or i==N-1:
                    con_mat[i][j] = 1
                else:
                    con_mat[i][j] = 2
            elif j==i-1 or j==i+1:
                con_mat[i][j] = -1
    return con_mat
        
#print(chain(10))        
evalue,evec=np.linalg.eig(chain(10))
for i in range(len(evalue)):
    print(evalue[i])
```

```python
def Veff(x):
    return (1.0-epsilon/2.0)*(tau/2.0)*2.0*fs(x,v0,L)*fs_grad(x,v0,L) - vw
    #return (1.0-epsilon1/2.0)*(tau/2.0)*(1/((1+q)**2))*2.0*fs(x,v0,L)*fs_grad(x,v0,L) - vw


def Deff(x):
    return Dt + (tau/2.0)*((fs(x,v0,L))**2)
    #return D/(1+q) + (tau/2.0)*((fs(x,v0,L))**2)*(1/((1+q)**2))

def b_func(x):
    return ((1.0-epsilon/2.0)*(tau/2.0)*2.0*fs(x,v0,L)*fs_grad(x,v0,L) - vw)/(Dt + (tau/2.0)*((fs(x,v0,L))**2))
    #return ((1.0-epsilon1/2.0)*(tau/2.0)*(1/((1+q)**2))*2.0*fs(x,v0,L)*fs_grad(x,v0,L) - vw)/(D/(1+q) + (tau/2.0)*((fs(x,v0,L))**2)*(1/((1+q)**2)))

    
def steady_state_density(x):
    def density_inner_int(y):
        g1 = x
        g2 = x+y
        return np.exp(-quad(b_func,g1,g2,limit=60)[0])
    return (1/(Dt + (tau/2.0)*((fs(x,v0,L))**2)))*quad(density_inner_int,0,L,limit=60)[0]
    #return (1/(D/(1+q) + (tau/2.0)*((fs(x,v0,L))**2)*(1/((1+q)**2))))*quad(density_inner_int,0,L,limit=60)[0]



def outer_int():
    "integrates over x from 0 to L"
    def middle_int(x):
        "integrates over y from 0 to L to give a function of x"
        def inner_int(y):
            "integrates over y' from x to x+y to give a function of x and y"
            h1 = x
            h2 = x+y
            return np.exp(-quad(b_func,h1,h2,limit=60)[0])
        return (1/(Dt + (tau/2.0)*((fs(x,v0,L))**2)))*quad(inner_int,0,L,limit=60)[0]
        #return (1/(D/(1+q) + (tau/2.0)*((fs(x,v0,L))**2)*(1/((1+q)**2))))*quad(inner_int,0,L,limit=60)[0]
    return quad(middle_int,0,L,limit=60)[0]
```

```python
def density_pol(x):
    return (1 + (tau/(2*Dt))*(fs(x,v0,L)**2))**(-(.5*epsilon))
```

```python
def density_stiff(x):
    epsilon1 = 1 - (q**2/(q + (1+q)*(tau*k/gamma)))
    #epsilon1 = 1 - q/2
    return (1 + (1/(1+q))*(tau/(2*T*gamma))*(fs(x,v0,L)**2))**(-(.5*epsilon1))
```

```python
%%time

L = 20.
gamma = 1.
T = 0.001
Dr = 10.
tau = 1/Dr
k = 5.
mu = 1.0
v0=1.0
Dt = 0.001
D=0.001
vw=0.25
xx = np.linspace(0,L,499)

tmp2 = []
tmp2 = np.vstack([xx])

fig, axs = plt.subplots(2, 1, sharex=True,height_ratios=[4,1])
fig.subplots_adjust(hspace=0)

for N in [2,5,8]:
#for q in [1,4]:

    chain(N)
    evalue,evec=np.linalg.eig(chain(N))
    summ=0
    for i in range(len(evalue)):
        if evalue[i]<10**(-9):
            evalue[i]=0
        #print(evalue[i])
        summ += 1/(1+tau*k*mu*evalue[i])
        #print(summ)

    epsilon = 2.0 - summ
    #epsilon1 = 1 - (q**2/(q + (1+q)*(tau*k/gamma)))
    print(epsilon)

    norm_pol = L/(quad(density_pol,0.0,L)[0])
    #norm_stiff = L/(quad(density_stiff,0.0,L)[0])
    cons1 = outer_int()

    axs[0].plot(xx,[L*steady_state_density(i)/(cons1) for i in xx],label='Numerical, N = %.2f' %N)
    axs[0].plot(xx,[density_pol(i)*norm_pol for i in xx],ls='--',label='Analytical, N = %.2f' %N)
    tmp2 = np.vstack([tmp2,[L*steady_state_density(i)/(cons1) for i in xx]])
    axs[0].set_ylabel(r"$\frac{\rho(x)}{\rho_b}$",fontsize=18,labelpad=20,rotation='horizontal')
    axs[0].set_xlabel(r"$x$")
    axs[0].legend(loc="best", fontsize=10, frameon=True)

    axs[1].plot(xx,fs(xx,v0,L),'k-')
    axs[1].set_xlabel(r"$x$",fontsize=18)
    axs[1].set_ylabel(r"$f_s(x)$",fontsize=15,labelpad=20,rotation='horizontal')
```

## Steady-state density in the comoving frame

The steady-state density in the comoving frame is given by (add proof to paper):

$$\frac{\rho(\chi_0)}{\rho_b} = \frac{L\mathcal{D}_{\text{eff}}^{-1}(\chi_0)\int_{0}^{L}dx\,\exp{\big\{-\int_{\chi_0}^{\chi_0+x}dy\,\frac{\mathcal{V}_{\text{eff}}(y)}{\mathcal{D}_{\text{eff}}(y)}\big\}}}{\int_{0}^{L}du\,\int_{0}^{L}\,\mathcal{D}_{\text{eff}}^{-1}(u)\exp{\big\{-\int_{u}^{u+x}dy\,\frac{\mathcal{V}_{\text{eff}}(y)}{\mathcal{D}_{\text{eff}}(y)}\big\}}}$$

In the case of polymers in traveling activity waves, the functions $\mathcal{V}_{\text{eff}}$ and $\mathcal{D}_{\text{eff}}$ are:
$$\mathcal{V}_{\text{eff}}(\chi_0) = \bigg(1-\frac{\epsilon}{2}\bigg)\frac{\tau}{d}\nabla_0\,v_a^2(\frac{\chi_0}{\sqrt{N}}) - v_w$$

$$\mathcal{D}_{\text{eff}} = D + \frac{\tau}{d}v_a^2(\frac{\chi_0}{\sqrt{N}})$$

where $\epsilon$ is given by:
$$\epsilon = 1 - \sum_{i=1}^{N-1}\frac{1}{1+\tau\mu\kappa\lambda_i}$$

```python
%%time

N = 8
Dr = 10.
tau = 1/Dr
k = 5.
mu = 1.0

chain(N)
evalue,evec=np.linalg.eig(chain(N))
summ=0
for i in range(len(evalue)):
    if evalue[i]<10**(-9):
        evalue[i]=0
    #print(evalue[i])
    summ += 1/(1+tau*k*mu*evalue[i])
    print(summ)

epsilon = 2.0 - summ
print(epsilon)
```
